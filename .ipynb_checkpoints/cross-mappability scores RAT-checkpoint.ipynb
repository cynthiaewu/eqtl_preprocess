```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
```

```python
cross_mappability_scores = pd.read_csv("cross_mappability.txt", sep="\t", names=['Gene1', 'Gene2', 'Cross-Mappability-score'])
cross_mappability_scores
```

```python
sns.histplot(data=cross_mappability_scores, x="Cross-Mappability-score", log_scale=True, kde=True)
```

```python
sns.histplot(data=cross_mappability_scores, x="Cross-Mappability-score", bins=8, binwidth=10)
```

```python
grouped_df = cross_mappability_scores.groupby(by=['Cross-Mappability-score']).size().reset_index(name = 'counts')
grouped_df
```

```python
filtered_df_scores_1 = cross_mappability_scores.loc[cross_mappability_scores['Cross-Mappability-score']<80]
filtered_df_scores_1
```

```python
grouped_f_1 = filtered_df_scores_1.groupby(by=['Cross-Mappability-score']).size().reset_index(name = 'counts')
grouped_f_1
```

```python
sorted_grouped_df = grouped_df.sort_values(by='counts')
sorted_grouped_df
```

```python
filtered_grouped = sorted_grouped_df.loc[sorted_grouped_df['counts']>1000]
filtered_grouped
```

```python
filtered_grouped.size
```

```python
filtered_grouped.max()
```

```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
 
# define the true objective function
def objective(x, a, b, c):
 return a * 1/x + b * x**2 + c
 
# choose the input and output variables
x, y = filtered_grouped['Cross-Mappability-score'], filtered_grouped['counts']
# curve fit
popt, _ = curve_fit(objective, x, y)
# summarize the parameter values
a, b, c = popt
print('y = %.5f * x + %.5f * x^2 + %.5f' % (a, b, c))
# plot input vs output
plt.scatter(x, y)
# define a sequence of inputs between the smallest and largest known inputs
x_line = np.arange(min(x), max(x), 1)
# calculate the output for the range
y_line = objective(x_line, a, b, c)
# create a line plot for the mapping function
plt.plot(x_line, y_line, '--', color='red')
plt.show()
```

```python
model1 = np.poly1d(np.polyfit(x, y, 1))
model2 = np.poly1d(np.polyfit(x, y, 2))
model3 = np.poly1d(np.polyfit(x, y, 3))
model4 = np.poly1d(np.polyfit(x, y, 4))
model5 = np.poly1d(np.polyfit(x, y, 5))

#create scatterplot
polyline = np.arange(min(x), max(x), 1)
plt.scatter(x, y)

#add fitted polynomial lines to scatterplot 
plt.plot(polyline, model1(polyline), color='green')
plt.plot(polyline, model2(polyline), color='red')
plt.plot(polyline, model3(polyline), color='purple')
plt.plot(polyline, model4(polyline), color='blue')
plt.plot(polyline, model5(polyline), color='orange')
plt.show()
```
